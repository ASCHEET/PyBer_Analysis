```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "August", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as teh total fare in US dollars accumulted for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# 5.1.4 Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston', color='green', marker='D', linewidth=2,)
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# set the y limit between 0 and 45
plt.ylim(0, 45)
# create a title
plt.title("PyBer Fare by Month")
# add a legend
plt.legend()
plt.grid()
# https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.plot.html
# https://matplotlib.org/3.1.0/api/index.html#the-pyplot-api
# https://matplotlib.org/stable/tutorials/introductory/usage.html
```

```python
# Create the bar plot
plt.bar(x_axis, y_axis)
```

```python
plt.bar(x_axis, y_axis, color="green", label="Boston")
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title
plt.title("PyBer Fare by Month")
# Add legend
plt.legend()
```

```python
# Create horz plot
plt.barh(x_axis, y_axis)
```

```python
plt.barh(y_axis, x_axis)
```

```python
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# vertical bar chart 5.1.6
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
plt.bar(x_axis, y_axis, color="black", label='New York')
```

```python
# create a horz bar chart with ax.plt
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# create a horz bar chart with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
plt.legend()
```

```python
# 5.1.7 scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c="red", label='Chicago', s=y_axis)
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.legend()
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(y_axis, x_axis, s=y_axis_larger)
plt.ylabel("Date")
plt.xlabel("Fare($)")
#plt.legend()
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```
